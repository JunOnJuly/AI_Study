```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# AND 게이트
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    cal = np.sum(w*x) + b
    if cal <= 0:
        return 0
    else:
        return 1

print(f'AND(0, 0) => {AND(0, 0)}')
print(f'AND(0, 1) => {AND(0, 1)}')
print(f'AND(1, 0) => {AND(1, 0)}')
print(f'AND(1, 1) => {AND(1, 1)}')
```

```python
# OR 게이트
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    cal = np.sum(w*x) + b
    if cal <= 0:
        return 0
    else:
        return 1

print(f'OR(0, 0) => {OR(0, 0)}')
print(f'OR(0, 1) => {OR(0, 1)}')
print(f'OR(1, 0) => {OR(1, 0)}')
print(f'OR(1, 1) => {OR(1, 1)}')
```

# XOR 게이트
XOR 게이트 : 배타적 논리합 \
x1, x2 중 하나만 1일 경우 1을 출력 \
하나의 선형 방정식으로 표현 불가능 -> 단일 퍼셉트론으로는 표현 불가능 \
비선형 방정식으로 표현 -> 다중 퍼셉트론을 사용

```python
def make_graph():
    def temp_graph(grad, x1, y1):
        return y1 - x1*grad
    
    x1, x2 = np.random.rand(2)
    y1, y2 = np.random.rand(2)
    grad = (x1-x2) / (y1-y2)
    
    b = temp_graph(grad, x1, y1)
    x_index = np.array([-1, 2])
    y_index = x_index*grad + b
    return x_index, y_index
```

```python
plt.scatter([0, 1], [1, 0], c='b', label='true')
plt.scatter([0, 1], [0, 1], c='r', label='false')
plt.xlim(-1, 2)
plt.ylim(-1, 2)
for _ in range(4):
    x, y = make_graph()
    plt.plot(x, y, linestyle='--')
plt.title('XOR Gate Truth Table')
plt.legend()
plt.show()
```

```python
# XOR 게이트
def XOR(x1, x2):
    s1 = 1 - AND(x1, x2) #NAND 게이트
    s2 = OR(x1, x2)
    return AND(s1, s2)

print(f'XOR(0, 0) => {XOR(0, 0)}')
print(f'XOR(0, 1) => {XOR(0, 1)}')
print(f'XOR(1, 0) => {XOR(1, 0)}')
print(f'XOR(1, 1) => {XOR(1, 1)}')
```
